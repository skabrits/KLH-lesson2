# Краткая справка комманд

```python
%%bash

# В  <> указывается пояснение чем является аргумент команды, напр. <path> означает, что в команде на место <path>
#                                       нужно подставить путь в вашей файловой системе до интересующего вас объекта

pwd                                 # выводит текущее местоположение

cd <path>                           # перемещается в директорию по пути <path>

ls                                  # показывает файлы в текущей директории

ls -la <path>                       # показывает подробную информацию о всех файлах в текущей директории

chmod +rwx <path>                   # даёт всем права на чтоние запись и исполнение

mkdir -p <folder>                   # создаёт папку <folder>

touch <file>                        # создаёт пустой файл с именем <file>

mv <what [path]> <where [path]>     # перемещает из 1 в 2

cp <what [path]> <to where [path]>  # копирует из 1 в 2

rm -rf <path>                       # удаляет всё рекурсивно по пути

ln -sr <what [path]> <link name>    # создаёт символическую ссылку на 1 с именем 2

cat <file>                          # выводит содержимое файла <file>

cat > <file> <<EOF                  # записывает <your text> в файл <file>, если внутри текста есть исполняемые 
<your text>                         # в bash выражения -- выполняет их
EOF

cat >> <file> <<"EOF"               # не позволяет раскрывать переменные внутри <your text>
<your text>
EOF

tail -n <number of lines> <file>    # выводит последние <number of lines> линий из файла <file>

head -n <number of lines> <file>    # то же самое, но первые <number of lines>

head -с <number of инеуы> <file>    # то же самое, но первые <number of bytes> байт

/dev/null                           # "мусорка", всё записанное сюда исчезает бесследно

/dev/urandom                        # содержит бесконечный поток случайных чисел, у которого всё время 
                                    #                                                      обновляется начало файла

echo <your text>                    # выводит <your text> в консоль

echo -e <your text with \n>         # выводит <your text with \n>, учитывая действие спецсимволов, вроде \n

find <path> -maxdepth <depth of search> -name <name of file>  # ищет по пути <path> файлы с именем <name of file>
                                                              # не более, чем на <depth of search> папок вглубь

find <path> -maxdepth <depth of search> -name <name of file> -exec bash -c '<your comand>' \;  # для каждого
                                                                                               # найденного файла
                                                                                               # выполняет команду
                                                                                               # <your comand>
                                                                                               # чтобы использовать
                                                                                               # внутри команды
                                                                                               # путь до найденного
                                                                                               # файла, используйте 
                                                                                               # значок {}

"*<string>*"                                            # означает любую строку, содержащую текст <string> для grep

sed "s/<string to replace>/<string that replaces>/g"    # заменяет все вхождения 1 на 2

sed "3s/<string to replace>/<string that replaces>/g"   # заменяет 3 вхождение 1 на 2

grep "<string>" <file>                                  # ищет в файле <file> все вхождения строки <string>

grep -r "<string>" <path>                               # то же самое, но ищет по всем файлам в папке

"^<text>.*$"                                            # regex, подходит чтобы найти строку, содержащую <text>,
                                                        # ^ -- означает, что <text> должен быть вначале строки
                                                        # .* -- означает, что после <text> может идти любое
                                                        #                                 количество любых символов
                                                        # $ -- означает конец строки

<command1> | <command2> | <command3> > <file>           # пайплайн, направляет stdout <command1> в stdin <command2>,
                                                        #  а stdout <command2> в stdin <command3> и stdout 
                                                        #                       <command3> направляет в файл <file>
                                                        
<command1>; <command2>                                  # выполняет <command2> вне зависимости от результата
                                                        # <command1>

<command1> && <command2>                                # выполняет <command2> при успешном выполнении <command1>
```
